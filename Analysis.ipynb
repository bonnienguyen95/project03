```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
results = pd.read_csv("results.csv")
```

```python
results = pd.DataFrame(results)
nflx = results.loc[results['name'] == "NFLX"]
nflx = nflx[['high', 'hour']]
nflx
```

```python
fig1 = plt.figure(figsize=(12,5))
ax = sns.lineplot(data=nflx, x="hour", y="high").set_title("Netflix Highest Hourly Price")
plt.show()
```

```python
ddog = results.loc[results['name'] == "DDOG"]
ddog = ddog[['high', 'hour']]
ddog
```

```python
fig2 = plt.figure(figsize=(12,5))
ax = sns.lineplot(data=ddog, x="hour", y="high").set_title("Datadog Highest Hourly Price")
plt.show()
```
